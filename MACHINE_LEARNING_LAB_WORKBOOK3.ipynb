**11.Write a Python program to create dataframe using random data and
draw a plot (Clasification) without using sklearn**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(42)  # For reproducibility
num_samples = 100
feature_a = np.random.uniform(0, 10, num_samples)
feature_b = np.random.uniform(0, 10, num_samples)
target = np.where(feature_a + feature_b > 10, 1, 0)  # Simple classification rule

# Create a DataFrame
df = pd.DataFrame({'Feature_A': feature_a, 'Feature_B': feature_b, 'Target': target})

# Plot the data
plt.scatter(df['Feature_A'], df['Feature_B'], c=df['Target'], cmap='coolwarm', alpha=0.7)
plt.xlabel('Feature A')
plt.ylabel('Feature B')
plt.title('Data Distribution (Classification)')
plt.colorbar(label='Target')
plt.show()
```

**12.Write a Python program to create dataframe using random data and

**14.Write a Python program to create dataframe using random data and draw a plot for fiting data (predict)
between traing & testing data (Clasification) without using sklearn**

```python
#Creating a DataFrame with Random Data:
import pandas as pd
import numpy as np

# Create a DataFrame with 10 rows and 3 columns of random integers
df = pd.DataFrame(np.random.randint(0, 100, size=(10, 3)), columns=list('ABC'))

# View the DataFrame
print(df)

#Splitting Data into Training and Testing Sets:
train_data = df.iloc[:8]
test_data = df.iloc[8:]


#Plotting the Data: Well create a scatter plot to visualize the distribution of the data points:
import matplotlib.pyplot as plt

plt.scatter(train_data['A'], train_data['B'], label='Training Data', color='blue')
plt.scatter(test_data['A'], test_data['B'], label='Testing Data', color='red')
plt.xlabel('Feature A')
plt.ylabel('Feature B')
plt.title('Data Distribution')
plt.legend()
plt.show()
```

**15.Write a Python program to load csv into dataframe and draw a plot for fiting data (predict) between traing &

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load CSV data into a DataFrame
csv_file = 'data.csv'  # Replace with your actual CSV file path
df = pd.read_csv(csv_file)

# Assuming your CSV has columns 'feature' and 'target'
# Split data into training and testing sets (80% training, 20% testing)
train_size = int(0.8 * len(df))
train_data = df.iloc[:train_size]
test_data = df.iloc[train_size:]

# Plot the data
plt.scatter(train_data['Pulse'], train_data['Maxpulse'], label='Training Data', color='blue')
plt.scatter(test_data['Pulse'], test_data['Maxpulse'], label='Testing Data', color='red')
plt.xlabel('Pulse')
plt.ylabel('Maxpulse')
plt.title('Data Distribution')
plt.legend()
plt.show()
```
